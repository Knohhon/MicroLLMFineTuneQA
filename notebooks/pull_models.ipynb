# Pull models from huggingface

---

### Load config

```python
import yaml

path = '../models_to_download.yaml'

with open(path, 'r', encoding='utf-8') as file:
    content = file.read().strip()
    config_data = yaml.safe_load(content)

config_data['models']
```

### Load models

```python
from huggingface_hub import snapshot_download
import os
from getpass import getpass

local_dir = config_data['local_dir']
os.makedirs(local_dir, exist_ok=True)
token = getpass('Enter the token from Huggingface')

for model_id in config_data['models']:

    snapshot_download(
        repo_id=model_id,
        local_dir=os.path.join(local_dir, model_id.replace("/", "--")),
        token=token
        )
```
