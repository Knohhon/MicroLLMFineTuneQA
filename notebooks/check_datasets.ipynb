```python
import pandas as pd
import json
```

```python
def open_jsonl_datasets(path):
    data = []
    with open(path, 'r', encoding='utf-8') as f:
        for line in f:
            p_data = json.loads(line)
            data.append(p_data)
    return data
```

```python
data_nq_dev_merge = open_jsonl_datasets(path='../datasets/NQ-open.dev.merged.jsonl')

len(data_nq_dev_merge)
```

```python
import gzip
import json

def open_gzip_jsonl_data(path):
    data = []
    with gzip.open(path, 'r') as f:
        for line in f:
            _ = json.loads(line)
            data.append(_)
    return data
```
